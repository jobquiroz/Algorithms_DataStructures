### Valid Palindrome

Given a string `s`, determine if it is a palindrome, considering only alphanumeric characters and ignoring cases.

#### Example 1
```
Input: s = "A man, a plan, a canal: Panama"
Output: true
Explanation: "amanaplanacanalpanama" is a palindrome.
```

#### Example 2
```
Input: s = "race a car"
Output: false
Explanation: "raceacar" is not a palindrome.
```

#### Constraints:

- `1 <= s.length <= 2 * 105`
- `s consists only of printable ASCII characters.`

````python
s1 = "A man, a plan, a canal: Panama"
s2 = "race a car"
````

````python
import string 

alphanum = string.ascii_lowercase + '1234567890'

alphanum
````

````python
def isPalindrome(s):
    # Removing non desired characters
    s = "".join([c.lower() for c in s if c.lower() in alphanum])
   # print(s)
    for i in range(len(s) // 2):  
    #    print(s[i], s[-i], i)
        if s[i] != s[-i-1]:
            return False
    return True
````

````python
isPalindrome(s1)
````

### Course's solution 3 (The pointer method)

````python
def isPalindrome(s):
    s = ''.join(filter(str.isalnum, s)).lower()
    
    # Defining pointers
    left, right = 0, len(s) - 1
    
    # When left = right or they pass the middle it breaks
    while left < right:
        if s[left] != s[right]:
            return False
        left = left + 1
        right = left - 1 
     
    return True
````

````python
isPalindrome(s1)
````

**Time complexity**: $O(n)$  (Actuaylly $O(n/2)$)

**Space complexity**: $O(1)$
